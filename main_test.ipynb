```python
from sklearn.datasets import make_classification
from model_functions import convert_lables
from model_functions import train_with_crossvalidation
from plotting import plot_decision_boundary
from SVM import SVM
```

```python
rand_state = 67 #seed for reproducibility (chenge to do more test on different databases)
X, Y = make_classification(n_samples= 1000, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=2, random_state=rand_state)
Y = convert_lables(Y)

#parameters  [can be tuned with a grid search]
splits = 5
C = 5
kernel = "poly"
degree = 3

#example of its functioning with poly kernel and a random dataset
svm = SVM(kernel=kernel, C = C, degree = degree, verbose = False)
svm, results = train_with_crossvalidation(splits, X, Y, svm)
print(" \n SVM mean accuracy score for %.d fold test in a 2D features environment: %.3f \n \n" % (splits, sum(results)/len(results))   )
plot_decision_boundary(svm,X,Y)
```
